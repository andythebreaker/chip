```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
max_of_int32=2147483647
min_of_int32=-2147483648
```

```python
adj_t=21

x = np.empty(1000000, dtype=np.float128)
x = np.linspace(min_of_int32, max_of_int32, num=1000000)

y = np.zeros_like(x)
for i in range(len(x)):
    z = abs(x[i]) if x[i] < 0 else x[i]
    if z <= (0.5 * 10**8):
        y[i] = x[i]*adj_t
    elif z >= (0.5 * 10**8) and z <= (1.2 * 10**8):
        y[i] = ((x[i] // 2) + (0.25 * 10**8) * np.sign(x[i]))*adj_t
    elif z >= (1.2 * 10**8) and z <= (2.4 * 10**8):
        y[i] = ((x[i] // 8) + (0.7 * 10**8) * np.sign(x[i]))*adj_t
    elif x[i] >= (2.4 * 10**8) and x[i] >= 0:
        y[i] = max_of_int32
    else:
        y[i] = min_of_int32

# Plotting the input and output
plt.plot(x, y)
plt.xlabel('Input (x)')
plt.ylabel('Output (y)')
plt.title('Input vs. Output')
plt.grid(True)
plt.show()
```
