```python
import numpy as np
import matplotlib.pyplot as plt

# Data
intervals = [[-5, -4], [-4, -3], [-3, -2], [-2, -1], [-1, 0], [0, 1], [1, 2], [2, 3], [3, 4], [4, 5]]
polynomials = [lambda x: 0.01129*x + 0.06248, lambda x: 0.02943*x + 0.13404, lambda x: 0.07177*x + 0.25602,
               lambda x: 0.14973*x + 0.41285, lambda x: 0.23105*x + 0.49653, lambda x: 0.23105*x + 0.50346,
               lambda x: 0.14973*x + 0.58714, lambda x: 0.07177*x + 0.74097, lambda x: 0.02943*x + 0.86595,
               lambda x: 0.01129*x + 0.93751]

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Plotting
fig, ax = plt.subplots()
ax.set_xlabel('x')
ax.set_ylabel('y')

for interval, polynomial in zip(intervals, polynomials):
    x = np.linspace(interval[0], interval[1], 100)
    y = polynomial(x)
    ax.plot(x, y, label=f'[{interval[0]}, {interval[1]}]: $P_1(x)$')

# Plotting sigmoid function
x_sigmoid = np.linspace(-5, 5, 100)
y_sigmoid = sigmoid(x_sigmoid)
ax.plot(x_sigmoid, y_sigmoid, label='Sigmoid')

ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

max_of_int32 = 2147483647
ctrl = (max_of_int32 // 2) // 5

# Data
intervals = [[-5 * ctrl, -4 * ctrl], [-4 * ctrl, -3 * ctrl], [-3 * ctrl, -2 * ctrl], [-2 * ctrl, -1 * ctrl],
             [-1 * ctrl, 0 * ctrl], [0 * ctrl, 1 * ctrl], [1 * ctrl, 2 * ctrl], [2 * ctrl, 3 * ctrl],
             [3 * ctrl, 4 * ctrl], [4 * ctrl, 5 * ctrl]]

polynomial_cases = {
    0: lambda x: 0.01129 * x + 0.06248 * ctrl,
    1: lambda x: 0.02943 * x + 0.13404 * ctrl,
    2: lambda x: 0.07177 * x + 0.25602 * ctrl,
    3: lambda x: 0.14973 * x + 0.41285 * ctrl,
    4: lambda x: 0.23105 * x + 0.49653 * ctrl,
    5: lambda x: 0.23105 * x + 0.50346 * ctrl,
    6: lambda x: 0.14973 * x + 0.58714 * ctrl,
    7: lambda x: 0.07177 * x + 0.74097 * ctrl,
    8: lambda x: 0.02943 * x + 0.86595 * ctrl,
    9: lambda x: 0.01129 * x + 0.93751 * ctrl
}

# Plotting
fig, ax = plt.subplots()
ax.set_xlabel('x')
ax.set_ylabel('y')

for interval in intervals:
    x = np.linspace(interval[0], interval[1], 100)
    interval_index = intervals.index(interval)
    polynomial = polynomial_cases.get(interval_index, None)
    if polynomial:
        y = polynomial(x)
        ax.plot(x, y, label=f'[{interval[0]}, {interval[1]}]: $P_1(x)$')

ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

max_of_int32=2147483647
min_of_int32=-2147483648

const_1 = 0.01129
const_2 = 0.06248 
const_3 = 0.02943
const_4 = 0.13404 
const_5 = 0.07177
const_6 = 0.25602
const_7 = 0.14973
const_8 = 0.41285 
const_9 = 0.23105
const_10 = 0.49653 
const_11 = 0.23105
const_12 = 0.50346
const_13 = 0.14973
const_14 = 0.58714 
const_15 = 0.07177
const_16 = 0.74097 
const_17 = 0.02943
const_18 = 0.86595
const_19 = 0.01129
const_20 = 0.93751

adj_t=10

x = np.empty(1000000, dtype=np.float128)
x = np.linspace(min_of_int32, max_of_int32, num=1000000)

ctrl=(max_of_int32//2)//5

y = np.zeros_like(x)
for i in range(len(x)):
    tmp=x[i]
    if tmp <= -5*ctrl:
        y[i] = 0#min_of_int32
    elif tmp >= -5*ctrl and tmp <= -4*ctrl:
        y[i] = (const_1*tmp + const_2*ctrl)*adj_t
    elif tmp >= -4*ctrl and tmp <= -3*ctrl:
        y[i] = (const_3*tmp + const_4*ctrl)*adj_t
    elif tmp >= -3*ctrl and tmp <= -2*ctrl:
        y[i] = (const_5*tmp + const_6*ctrl)*adj_t
    elif tmp >= -2*ctrl and tmp <= -1*ctrl:
        y[i] = (const_7*tmp + const_8*ctrl)*adj_t
    elif tmp >= -1*ctrl and tmp <= 0*ctrl:
        y[i] = (const_9*tmp + const_10*ctrl)*adj_t
    elif tmp >= 0*ctrl and tmp <= 1*ctrl:
        y[i] = (const_11*tmp + const_12*ctrl)*adj_t
    elif tmp >= 1*ctrl and tmp <= 2*ctrl:
        y[i] = (const_13*tmp + const_14*ctrl)*adj_t
    elif tmp >= 2*ctrl and tmp <= 3*ctrl:
        y[i] = (const_15*tmp + const_16*ctrl)*adj_t
    elif tmp >= 3*ctrl and tmp <= 4*ctrl:
        y[i] = (const_17*tmp + const_18*ctrl)*adj_t
    elif tmp >= 4*ctrl and tmp <= 5*ctrl:
        y[i] = (const_19*tmp + const_20*ctrl)*adj_t
    else:
        y[i] = max_of_int32

# Plotting the input and output
plt.plot(x, y)
plt.xlabel('Input (x)')
plt.ylabel('Output (y)')
plt.title('Input vs. Output')
plt.grid(True)
plt.show()
```
